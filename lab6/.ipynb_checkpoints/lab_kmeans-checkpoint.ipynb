# Кластерный анализ
## Метод к-средних



Дана матрица данных $X$ и дано число $k$ предполагаемых кластеров. Цель кластеризации представить данные в виде групп кластеров $C=\{C_1, C_2, \ldots, C_k\}$. Каждый кластер имеет свой центр:

\begin{equation}
\mu_i = \frac{1}{n_i} \sum \limits_{x_j \in C_i} x_j
\end{equation}

где $n_i = |C_i|$ - это количество точек в кластере $C_i$.

Таким образом, мы имеем некоторые кластеры $C=\{C_1, C_2, \ldots, C_k\}$ и необходимо оценить качество разбиения. Для этого будем вычислять сумму квадратических ошибок (SSE):

\begin{equation}
SSE(C) = \sum \limits_{i=1}^{k} \sum \limits_{x_j \in C_i} ||x_j - \mu_i||^2
\end{equation}

Цель найти
\begin{equation}
C^* = arg\min\limits_C \{SSE(C)\}
\end{equation}

### Алгоритм к-средних
На вход алгоритм получает матрицу данных $D$, количество кластеров $k$, и критерий остановки $\epsilon$:
1. t = 0
2. случайным образом инициализируем $k$ центров кластеров: $\mu_1^t, \mu_2^t, \ldots, \mu_k^t \in R^d$;
3. повторять
    4. $t = t + 1$;
    5. $C_j = 0$ для всех $j = 1, \ldots, k$
    6. для каждого $x_j \in D$
        7. $j^* = arg\min\limits_C \{||X_j - \mu_i^{t-1}||^2\}$ \\\ присваиваем $x_j$ к ближайшему центру
        8. $C_{j^*} = C_{j^*} \cup {x_j}$
    9. для каждого i=1 до k
        10. $\mu_i = \frac{1}{|C_i|} \sum_{x_j \in C_i} x_j$
11. пока $\sum_{i=1}^k ||\mu_i^{t} - \mu_i^{t-1}||^2 \leq \epsilon$

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import numpy as np
import random
%matplotlib inline

X, Y = make_blobs(n_samples = 1000, n_features=2, centers=5, cluster_std = 1.2, random_state=17)
plt.scatter(X[:,0], X[:,1], c=Y)
```

```python
# Поиск центра кластера
def get_cluster_center(cluster):
    x0_mean = cluster[:,0].mean()
    x1_mean = cluster[:,1].mean()
    return np.array([x0_mean, x0_mean])

# Нахождение квадратического расстояния между двумя точками
def dest_btw_points(point_1, point_2):
    return np.linalg.norm(point_1 - point_2)

#Разбиение массива X на кластеры по центрам
def drop_clusters(X, centers):
    clusters = dict()
        
    for point in X:           
        bestCluster = min([(index, dest_btw_points(point, center)) for index, center in enumerate(centers)],
                          key=lambda t:t[1])[0]
        try:
            clusters[bestCluster].append(point)
        except KeyError:
            clusters[bestCluster] = [point]
    return {key: np.array(clusters[key]) for key in clusters}


def reevaluateCenters(clusters):
    """Пересчет центров кластеров."""
    newCenters = list()
    for key in clusters:
        newCenters.append(np.mean(clusters[key], axis = 0))
    return np.array(newCenters)


def cluster_plot(clusters):
    """Построение графика кластеров."""
    for key in clusters:
        plt.scatter(clusters[key][0:,0], clusters[key][0:,1])
    plt.show()


def kMeans(X, k, early_stop, plot=False):
    """Реализация алгоритма."""
    centers = X[:k]
    eps = float("inf")
    while eps >= early_stop:
        clusters = drop_clusters(X, centers)
        if plot == True:
            cluster_plot(clusters)
        newCenters = reevaluateCenters(clusters)
        eps = sum([dest_btw_points(centers[index], newCenters[index]) for index, _ in enumerate(centers)])
        centers = newCenters
    return clusters, centers
```

```python
def sse(clusters, centers):
    result = 0
    for clusterId in clusters:
        for dot in clusters[clusterId]:
            result += dest_btw_points(dot, centers[clusterId])
    return result
```

```python
clusters, centers = kMeans(X, 5, 0.1, plot=True)
```

```python
for k in range(1, 11):
    clusters, centers = kMeans(X, k, 1.0)
    print("K = {}, SSE = {}".format(k, sse(clusters, centers)))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv', encoding='cp1251')
df = df.drop(columns=['Модель', 'Цвет'])
df.head()
```

```python
new_df = pd.get_dummies(df)
new_df.head()
```

```python
from sklearn.preprocessing import StandardScaler
stnd_scaler = StandardScaler()
new_df[['Год', 'Пробег', 'Объем', 'Мощность', 'Цена']] = stnd_scaler.fit_transform(new_df[['Год', 'Пробег', 'Объем', 'Мощность', 'Цена']])
new_df.head()
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, AffinityPropagation
inertia = []
for it in np.arange(10,11,1):
    method = KMeans(n_clusters=it)
    method.fit(new_df)
    inertia.append(method.inertia_)
    print(it)
```

```python
def cls_info(df, k):
    print('------ Кластер ', k, ' -------')
    claster =  df[method.labels_ == k]
    print(claster['Год'].mean())
    print(claster['Пробег'].mean())
    print(claster['Объем'].mean())
    print(claster['Мощность'].mean())
    print(claster['Цена'].mean())
    print(claster['Привод'].value_counts().head(1))
    print(claster['Марка'].value_counts().head(3))
    print(claster['Кузов'].value_counts().head(3))
    print(claster['КПП'].value_counts().head(1))
    print('---------------------------')
```
